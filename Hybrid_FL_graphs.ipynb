```python
# Importare le librerie necessarie
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline

### 3 ROUND
# asse x
ibridazione = ['0%', '20%', '40%', '60%', '80%', '100%', 'Single']
ibridazione_var = ['20%', '40%', '60%', '80%']
# asse y
tempo = [
    3040.670933924,
    2367.849591411,
    1877.749404199,
    1322.070100521,
    852.970493093,
    258.357760858,
    3229.329499484
]
tempo_var = [
    2367.849591411,
    1877.749404199,
    1322.070100521,
    852.970493093
]
min_tempo = [
    2329.653263528,
    1858.427478835,
    1313.817896305,
    826.751346579
]
max_tempo = [
    2425.823043402,
    1908.663467500,
    1330.150008355,
    915.776163148
]
scala_tempo = [500, 1000, 1500, 2000, 2500, 3000]

accuracy = np.array([
    0.351848509363846,
    0.758675391638012,
    0.873145838411949,
    0.923812413346023,
    0.954460679474149,
    0.969135802469135,
    0.140363062295577
]) * 100
accuracy_var = np.array([
    0.758675391638012,
    0.873145838411949,
    0.923812413346023,
    0.954460679474149
]) * 100
min_accuracy = np.array([
    0.738729324978393,
    0.867599606434557,
    0.912968316290575,
    0.940647124304267
]) * 100
max_accuracy = np.array([
    0.789793053321170,
    0.877640325042922,
    0.934009401451261,
    0.972382920110192
]) * 100

fscore = np.array([
    0.299649503226286,
    0.743052127402876,
    0.868718635756909,
    0.923017324722613,
    0.953297092116097,
    0.969115538695197,
    0.088826925412740
]) * 100
fscore_var = np.array([
    0.743052127402876,
    0.868718635756909,
    0.923017324722613,
    0.953297092116097
]) * 100
min_fscore = np.array([
    0.722549001832413,
    0.864001840260842,
    0.911839227279883,
    0.938168351530180
]) * 100
max_fscore = np.array([
    0.775886783136429,
    0.871680661939634,
    0.932419969029956,
    0.971575301789542
]) * 100

scala_perc = [20, 40, 60, 80, 90]

x = np.arange(len(ibridazione[:-1]))
x_long = np.linspace(x.min(), x.max(), 500)

spline_tempo = make_interp_spline(x, tempo[:-1])
y_tempo = spline_tempo(x_long)

spline_acc = make_interp_spline(x, accuracy[:-1])
y_acc = spline_acc(x_long)

spline_fs = make_interp_spline(x, fscore[:-1])
y_fs = spline_fs(x_long)

# plt configuration

###########
# TEMPO DI ESECUZIONE
#############
green_cmap = plt.cm.get_cmap('Greens', len(ibridazione)+1)
greens = green_cmap(np.arange(2, len(ibridazione)+1))

#barra senza variazione
plt.bar(ibridazione[0], tempo[0], capsize=10, color=greens[0], edgecolor='#303030')
# barre con variazione
plt.bar(ibridazione_var, tempo_var, yerr=[np.subtract(tempo_var, min_tempo), np.subtract(max_tempo, tempo_var)],
        capsize=10, color=greens[1:], edgecolor='#303030', error_kw=dict(lw=1.5, capsize=10, capthick=1.5))
#barre senza variazione
plt.bar([ibridazione[5],ibridazione[6]], [tempo[5],tempo[6]],
        capsize=10, color=[greens[-1],'darkgray'], edgecolor='#303030')
#plt.plot(x_long, y_tempo, color='black', marker=',')
for valore in scala_tempo:
    plt.axhline(valore, color='gray', linestyle='dashed', linewidth=0.5)

plt.xlabel('Hybridization')
plt.ylabel('Time (s)')
plt.title("Execution time", fontsize = 15)

plt.show()

######
# ACCURACY
######
blue_cmap = plt.cm.get_cmap('Blues', len(ibridazione)+1)
blues = blue_cmap(np.arange(2,len(ibridazione)+1))

#barra senza variazione
plt.bar(ibridazione[0], accuracy[0], capsize=10, color=blues[0], edgecolor='#303030')
# barre con variazione
plt.bar(ibridazione_var, accuracy_var, yerr=[np.subtract(accuracy_var, min_accuracy), np.subtract(max_accuracy, accuracy_var)],
        capsize=10, color=blues[1:], edgecolor='#303030', error_kw=dict(lw=1.5, capsize=10, capthick=1.5))
#barre senza variazione
plt.bar([ibridazione[5],ibridazione[6]], [accuracy[5],accuracy[6]],
        capsize=10, color=[blues[-1],'darkgray'], edgecolor='#303030')
#plt.plot(x_long, y_acc, color='black', marker=',')
plt.ylim(0, 100)
for valore in scala_perc:
    plt.axhline(valore, color='gray', linestyle='dashed', linewidth=0.5)

plt.xlabel('Hybridization')
plt.ylabel('Accuracy (%)')
plt.title("Accuracy", fontsize = 15)

plt.show()

#######
# F-SCORE
#######
reds_cmap = plt.cm.get_cmap('Reds', len(ibridazione)+1)
reds = reds_cmap(np.arange(2,len(ibridazione)+1))

#barra senza variazione
plt.bar(ibridazione[0], fscore[0], capsize=10, color=reds[0], edgecolor='#303030')
# barre con variazione
plt.bar(ibridazione_var, fscore_var, yerr=[np.subtract(fscore_var, min_fscore), np.subtract(max_fscore, fscore_var)],
        capsize=10, color=reds[1:], edgecolor='#303030', error_kw=dict(lw=1.5, capsize=10, capthick=1.5))
#barre senza variazione
plt.bar([ibridazione[5],ibridazione[6]], [fscore[5],fscore[6]],
        capsize=10, color=[reds[-1],'darkgray'], edgecolor='#303030')
#plt.plot(x_long, y_fs, color='black', marker=',')
plt.ylim(0, 100)
for valore in scala_perc:
    plt.axhline(valore, color='gray', linestyle='dashed', linewidth=0.5)

plt.xlabel('Hybridization')
plt.ylabel('F-Score (%)')
plt.title("F-Score", fontsize = 15)

plt.show()
```
